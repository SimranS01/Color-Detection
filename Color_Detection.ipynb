```python
import numpy as np
import cv2
import pandas as pd
from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')  

def pick_image():
    img_name= input('Write the image file name here:')
    image= cv2.imread(img_name)
    return image

index=['COLOR','COLOR_NAME','HEXA','R','G','B']
csv=pd.read_csv('colors.csv',names=index,header=None)

def color_finder(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        global b,g,r,clicked
        clicked= True
        b,g,r= img[y,x]
        b=int(b)
        g=int(g)
        r=int(r)

def find_color_name(R,G,B):
    minimum= 9999
    for i in range(len(csv)):
        data= abs(R-int(csv.loc[i,"R"]))+abs(G-int(csv.loc[i,"G"]))+ abs(B-int(csv.loc[i,"B"]))
        if(data<=minimum):
            minimum=data
            colorname=csv.loc[i,'COLOR_NAME']
    return colorname

def live_pic():
    webcam = cv2.VideoCapture(0)
    while True:
        ret, img = webcam.read()
        cv2.imshow("DETECTION", img)
        key = cv2.waitKey(125)
        if key == ord('s'): 
            cv2.imwrite(filename='saved.jpg', img=img)
            webcam.release()
            imgn = cv2.imread('saved.jpg')
            cv2.imshow('DETECTION', imgn)
            cv2.waitKey(1)
            print('Captured')
            break
            cv2.destroyAllWindows()
        elif key == ord('q'):
            print("Turning off camera.")
            webcam.release()
            print("Camera off.")
            print("Program ended.")
            cv2.destroyAllWindows()
            break
            
choice=0
while choice not in range(1,3):
    choice=int(input('press 1 if you want to find colors in a downloaded picture\n press 2 if you want to find colors on a live photo'))
    if choice not in range(1,3):
        print('The entered value is unacceptable. please choose either 1 or 2')
if choice == 1:
    clicked = False
    img = pick_image()
    cv2.namedWindow('DETECTION',cv2.WINDOW_NORMAL)
    cv2.setMouseCallback('DETECTION',color_finder)
    while True:
        cv2.imshow("DETECTION",img)
        if clicked==True:
            cv2.rectangle(img,(1,5),(270,35),(b,g,r),-1)
            text= find_color_name(r,g,b)
            cv2.putText(img,text,(1,25),2,0.8,(255,255,255),2,cv2.FONT_HERSHEY_COMPLEX)
            if(r+g+b>=600):
                cv2.putText(img,text,(1,25),2,0.8,(0,0,0),2,cv2.FONT_HERSHEY_COMPLEX)
            clicked=False
        k=cv2.waitKey(10)
        if k & 0xFF == 27:
            break
    cv2.destroyAllWindows()
else:
    live_pic()
    img= cv2.imread('saved.jpg')
    cv2.namedWindow('DETECTION',cv2.WINDOW_NORMAL)
    cv2.setMouseCallback('DETECTION',color_finder)
    clicked=False
    while True:
        cv2.imshow("DETECTION",img)
        if clicked==True:
            cv2.rectangle(img,(1,5),(270,30),(b,g,r),-1)
            text= find_color_name(r,g,b)
            cv2.putText(img,text,(1,25),2,0.8,(255,255,255),2,cv2.FONT_HERSHEY_COMPLEX)
            if(r+g+b>=600):
                cv2.putText(img,text,(1,25),2,0.8,(0,0,0),2,cv2.FONT_HERSHEY_COMPLEX)
            clicked=False
        k=cv2.waitKey(10)
        if k & 0xFF ==27:
            break
    cv2.destroyAllWindows()
```
